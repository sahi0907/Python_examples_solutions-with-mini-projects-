```python
#make a simple calculator
def calculator(a,b,operator):
  if operator == '+':
    return a+b
  elif operator == '-':
    return a-b
  elif operator == '*':
    return a*b
  elif operator == '/':
    return a/b
  else:
    return "invalid operator"
calculator(588,617,'*')
```

```python
def apply_discount(total):
  if total >= 5000:
    return total * 0.90
  elif total >=2000 :
    return total *0.95
  else:
    return total
apply_discount(2050)
```
